```python
import string

import nltk
from tokenize import tokenize
from razdel import sentenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

file = open('new 5.txt', encoding='utf8')
text = file.read()
sents_nltk = sent_tokenize(text, language="russian")
print(sents_nltk)

tokens_nltk = [word for sent in sents_nltk for word in word_tokenize(sent, language="russian")]
tokens = list(filter(lambda word: word not in string.punctuation and word.isalpha, tokens_nltk))
print(tokens)
print(f"count {len(tokens)}")
print(f"unique {len(set(tokens))}")

# your code here
```
